Create a series `star_ratings` with the number of stars corresponding to each review in the dataset.

```python
#복잡한 점수를 간단한 별 등급으로 번역. 95점 이상은 3점, 85점 이상 95점 미만은 2점 부여. 다른 점수는 별 1개입니다.
#캐나다 와인은 포인트에 관계없이 자동으로 별 3개 부여
#데이터 세트의 각 리뷰에 해당하는 별 개수로 star_ratings 시리즈 만들기

def star(x):
    if x['country'] == 'Canada':
        return 3
    elif x['points'] >= 95 :
        return 3
    elif x['points'] >= 85 :
        return 2
    else :
          return 1
star_ratings = reviews.apply(star, axis =1)
# axis= 0(index) : 행에 따라 동작
# axis= 1(columns): 열에 따라 동작

# Check your answer
q7.check()
```
