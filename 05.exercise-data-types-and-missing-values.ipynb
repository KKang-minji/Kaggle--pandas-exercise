## 4.
What are the most common wine-producing regions? Create a Series counting the number of times each value occurs in the `region_1` field. This field is often missing data, so replace missing values with `Unknown`. Sort in descending order.  Your output should look something like this:

```
Unknown                    21247
Napa Valley                 4480
                           ...  
Bardolino Superiore            1
Primitivo del Tarantino        1
Name: region_1, Length: 1230, dtype: int64
```

와인을 가장 많이 생산하는 지역은 어디인가요? region_1 필드에서 각 값이 발생하는 횟수를 세는 Series를 만듭니다. 이 필드는 데이터가 누락되는 경우가 많으므로 누락된 값을 알 수 없음으로 대체하십시오. 내림차순으로 정렬합니다. 출력은 다음과 같아야 합니다.

**fillna(): 누락된 값을 바꾸는 것은 일반적인 작업**

.fillna("Unknown") => 결측값을 "unknown"으로 바꿈

**value_counts: 종류별 갯수 세기**

**sort_values: 데이터 값을 기준으로 정렬(ascending=True/False 오름차순/내림차순)**

````python
reviews_per_region = reviews.region_1.fillna("Unknown").value_counts().sort_values(ascending=False)

# Check your answer
q4.check()
reviews_per_region
````
